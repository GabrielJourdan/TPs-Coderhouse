```python
import numpy as np

#1. Generar un array aleatorio de 100 elementos. Calcular la Mediana correspondiente:

np_random_dim = np.random.randint(150,size=(100))
print(np_random_dim)

np.median(np_random_dim)
```

```python
#2. Generar un array que contenga siempre valor 1 en sus elementos y que tenga 20 elementos, luego otro array de rango 1 hasta el 20 inclusive. Sumarlos y luego decirme cuál es el Promedio del array resultante:

np_relleno = np.full(20,1)
print(np_relleno)

np_rango = np.arange(1,21)
print(np_rango)

np_suma = np_relleno + np_rango
print(np_suma)

np_suma.mean()
```

```python
#3. En base al array del ajedrez generado en esta diapositiva, hacer un cambio de forma (reshape), y que sea de 4 filas x 16 columnas. Calcular su Desvío Estándar

np.arange(1,65)
print(np.arange(1,65))

np_new = np.arange(1,65).reshape(8,8)
print(np_new)

np_new2 = np_new.reshape(4,16)
print(np_new2)

np.std(np_new2)

#Nota: El original es de 1 x 65, y luego pide cambiarlo a 8 x 8. Por eso hice ambos pasos.
```
