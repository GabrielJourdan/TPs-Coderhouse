```python
#4. Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.

indice = list(df_jjoo.Year)
columnas = df_jjoo.columns[6:7]
valores = df_jjoo.values[:,6:7]
df_jjoo_medal = pd.DataFrame(data = valores , index= indice, columns = columnas)

df_jjoo_medal.sum(axis= 'columns')

df_men = df_jjoo_medal[df_jjoo_medal['Gender'] == 'Men']
df_men_medals = pd.DataFrame(df_men.index.value_counts()).reset_index()
df_men_medals.columns = ['Año','Medals Mens']
df_men_medals
```
