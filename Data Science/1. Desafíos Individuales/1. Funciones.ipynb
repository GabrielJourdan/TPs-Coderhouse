```python
#1. Escribir una función para calcular el factorial de 10:

def factorial (n):
    inicial = 1
    for x in range(1,n+1):
      inicial *= x
    return inicial
print(factorial(10))
```

```python
#2. Escribir una función para calcular la suma de un rango comenzando en 10 y terminando en 25 incluído. Ese rango debe ir de uno en uno:

def suma (n):
    inicial = 0
    for i in range(10,n+1):
      inicial = inicial + i
    return inicial
print(suma(25))
```
