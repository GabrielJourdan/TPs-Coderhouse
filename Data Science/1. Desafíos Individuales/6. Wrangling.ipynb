```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

###2. Corregir los Target a solo dos valores distintos ‘<=50k’ y ‘>50k’

Traemos los valores de la columna Target para ver cuáles debemos corregir:

```python
df_Total.Target.unique()
```

```python
def corregir_target(x):
   if x in [' <=50K', ' <=50K.']: 
    return '<=50k'
   elif x in [' >50K', ' >50K.']: 
    return '>50k'
   else:
    return x
```

```python
df_Total['New_Target'] = df_Total.Target.apply(lambda x: corregir_target(x))
```

```python
df_Total['New_Target']= df_Total.Target.transform(lambda x: corregir_target(x) )

df_Total
```

###Revisamos que hayan quedado correctamente generados los valores únicos solicitados: ‘<=50k’ y ‘>50k’

```python
df_Total.New_Target.unique()
```

###3. Calcular la proporción que hay entre personas con salarios >50k respecto del total entre personas de distinto Sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y educación


Creamos una nueva columna para contar la cantidad de personas por fila:

```python
from numpy.ma.core import count
df_Total['Personas'] = df_Total['ID'].apply(lambda x: count(x))
df_Total.head()
```

Hacemos un Groupby para armar la tabla comparativa por Sexo:

```python
df_group = df_Total.groupby(['Sex', 'New_Target']).agg(sum_personas=('Personas', 'sum'))
df_group
```

Hacemos la tabla para el cálculo del promedio:

```python
df_group['sum_personas'].groupby(level=0).apply(lambda x : (x / float(x.sum())))
```

####Repetimos ambos procedimientos para Raza

Hacemos un Groupby para armar la tabla comparativa por Raza:

```python
df_group = df_Total.groupby(['Race', 'New_Target']).agg(sum_personas=('Personas', 'sum'))
df_group
```

Hacemos la tabla para el cálculo del promedio:

```python
df_group['sum_personas'].groupby(level=0).apply(lambda x : (x / float(x.sum())))
```

####Repetimos ambos procedimientos para Educación

Hacemos un Groupby para armar la tabla comparativa por Educación:

```python
df_group = df_Total.groupby(['Education', 'New_Target']).agg(sum_personas=('Personas', 'sum'))
df_group
```

Hacemos la tabla para el cálculo del promedio:

```python
df_group['sum_personas'].groupby(level=0).apply(lambda x : (x / float(x.sum())))
```

Las mujeres que cobran por encima de los 50K son solo el 11%, mientras que los hombres que cobran esos salarios alcanzan el 30%.

Las personas del Pacífico Asiático gananan más en promedio que las del resto de las razas, llegando a un 27% de personas con salarios superiores a 50K, mientras que la raza que le sigue es la blanca (25%), y el resto promedia el 12%.

Con respecto al nivel de Educación, observamos que a mayor avance en las carreras, mayores son los salarios que superan los 50K, siendo los de las personas con doctorados los que alcanzan el mayor porcentaje: 72%
